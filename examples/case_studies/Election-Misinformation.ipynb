# Election Misinformation OSINT Triage

```python
from enum import Enum

class EmotionalTone(Enum):
    NEUTRAL = 'neutral'
    BIASED = 'biased'
    MANIPULATIVE = 'manipulative'

def assess(text: str) -> EmotionalTone:
    lowered = text.lower()
    if any(k in lowered for k in ['must', 'guarantee', 'secret', 'unbelievable', 'shock']):
        return EmotionalTone.MANIPULATIVE
    if any(k in lowered for k in ['clearly', 'obviously', 'undoubtedly']):
        return EmotionalTone.BIASED
    return EmotionalTone.NEUTRAL
```

Evaluate sample posts.

```python
posts = [
    'Officials guarantee the election was rigged.',
    'Preliminary counts were released on Monday.',
]
[assess(p) for p in posts]
```
