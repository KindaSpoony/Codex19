# Critical Infrastructure Threat Escalation

```python
def integrate_tone_feedback(answer: str) -> str:
    if 'must' in answer.lower():
        return answer + "\n\n> Tone-Check: Source material shows *manipulative* language – interpret with caution."
    return answer
```

Demonstrate disclosure.

```python
statement = 'You must act now to disrupt the facility.'
integrate_tone_feedback(statement)
```
